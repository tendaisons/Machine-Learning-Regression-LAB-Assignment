```python
#QUESTION 5:Multiple Linear Regression with Categorical Variables
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error
from sklearn.preprocessing import OneHotEncoder

# Generate synthetic data
np.random.seed(42)
n_samples = 300

square_footage = np.random.randint(1000, 3000, n_samples)
num_bedrooms = np.random.randint(2, 6, n_samples)
neighborhoods = np.random.choice(['A', 'B', 'C'], n_samples)

# Create a DataFrame
data = pd.DataFrame({
    'SquareFootage': square_footage,
    'NumBedrooms': num_bedrooms,
    'Neighborhood': neighborhoods
})

# Display all rows of the dataset
pd.set_option('display.max_rows', None)
print(data)


# One-hot encode the categorical variable (Neighborhood)
data = pd.get_dummies(data, columns=['Neighborhood'])

# Generate house prices based on the features
data['HousePrice'] = 1000*data['SquareFootage'] + 5000*data['NumBedrooms'] + \
                     20000*data['Neighborhood_A'] + 15000*data['Neighborhood_B'] + 10000*data['Neighborhood_C'] + \
                     np.random.normal(0, 50000, n_samples)

# Split data into features and target variable
X = data.drop('HousePrice', axis=1)
y = data['HousePrice']

# Fit a Multiple Linear Regression model
model = LinearRegression()
model.fit(X, y)

# Make predictions
y_pred = model.predict(X)

# Evaluate the model
r2 = r2_score(y, y_pred)
mse = mean_squared_error(y, y_pred)

print("R-squared:", r2)
print("Mean Squared Error:", mse)



# Plot actual vs. predicted house prices
plt.figure(figsize=(10, 6))
plt.scatter(range(len(y)), y, color='blue', label='Actual House Prices')
plt.scatter(range(len(y)), y_pred, color='red', label='Predicted House Prices')
plt.title('Actual vs. Predicted House Prices')
plt.xlabel('Sample Index')
plt.ylabel('House Price')
plt.legend()
plt.show()

print("""
INTERPRETATION OF RESULTS

R-squared:

The R-squared value is 0.9918, which indicates that approximately 99.18% of the variance in house prices is explained by the features included in the model.
This high R-squared value suggests that the model fits the data very well and can effectively predict house prices based on square footage, number of bedrooms, and neighborhood.
Mean Squared Error (MSE):

The Mean Squared Error is 2574079899.46.

MSE represents the average squared difference between the actual house prices and the predicted house prices.

In this case, the relatively low MSE indicates that the model's predictions are close to the actual house prices on average.

In summary, the high R-squared value and low MSE suggest that the Multiple Linear Regression model is performing very well in predicting house prices based on the provided features. The model explains a significant amount of the variance in house prices and makes accurate predictions with relatively small errors.""")


#NOTES
#We generate synthetic data for square footage, number of bedrooms, and neighborhood.
#One-hot encode the categorical variable (neighborhood).
#Create house prices based on the features.
#Fit a Multiple Linear Regression model to predict house prices.
#Evaluate the model's accuracy using R-squared and Mean Squared Error.
#You can further analyze the coefficients of the model to interpret the impact of each feature on the house prices. Feel free to adjust the data generation process and model parameters to explore different scenarios and improve the model's accuracy.
```
