```python
# QUESTION 1:Linear Regression
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt


# Set the random seed for reproducibility
np.random.seed(42)

# Generate the number of hours studied (X) with a range of 1 to 10
hours_studied = np.random.uniform(1, 10, 50)

# Generate the corresponding exam scores (Y) using a linear relationship with noise
true_coefficient = 5
true_intercept = 20
noise = np.random.normal(0, 5, 50)
exam_scores = true_intercept + true_coefficient * hours_studied + noise

# Create a DataFrame to store the dataset
dataset = pd.DataFrame({'Hours Studied': hours_studied, 'Exam Scores': exam_scores})

# Display all rows of the dataset
print(dataset)


# Assuming 'dataset' is the DataFrame containing your data
X = dataset[['Hours Studied']]  # Features (number of hours studied)
y = dataset['Exam Scores']     # Target variable (exam scores)

# Fit the linear regression model
model = LinearRegression()
model.fit(X, y)

# Make predictions
y_pred = model.predict(X)

# Evaluate the model
mse = mean_squared_error(y, y_pred)
r2 = r2_score(y, y_pred)

# Print the coefficients
print('Intercept:', model.intercept_)
print('Coefficient:', model.coef_[0])

# Print the evaluation metrics
print('Mean Squared Error:', mse)
print('R-squared:', r2)

# Plot the regression line
plt.scatter(X, y, color='blue')
plt.plot(X, y_pred, color='red', linewidth=2)
plt.xlabel('Hours Studied')
plt.ylabel('Exam Scores')
plt.title('Linear Regression - Hours Studied vs Exam Scores')
plt.show()
```
